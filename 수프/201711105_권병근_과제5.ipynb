<p class="admonition-title">연습 문제 4.4.4</p>
<p>타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 <code class="docutils literal notranslate"><span class="pre">category1</span></code> 열을 만들어라. <code class="docutils literal notranslate"><span class="pre">category1</span></code> 카테고리는 다음과 같이 정의된다.</p>
<ol class="simple">
<li><p>20살이 넘으면 성별을 그대로 사용한다.</p></li>

```python
titanic['category1'] = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
titanic
```

<p class="admonition-title">연습 문제 4.4.5</p>
<p>타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.</p>
</div>
</div>

```python
mean_age = titanic["age"].mean()
titanic["age"] = titanic["age"].fillna(mean_age)
titanic
```

<p class="admonition-title">연습 문제 4.4.6</p>
<p>타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 <code class="docutils literal notranslate"><span class="pre">category2</span></code> 열을 만들어라. <code class="docutils literal notranslate"><span class="pre">category2</span></code> 카테고리는 다음과 같이 정의된다.</p>
<ol class="simple">
<li><p>성별을 나타내는 문자열 <code class="docutils literal notranslate"><span class="pre">male</span></code> 또는 <code class="docutils literal notranslate"><span class="pre">female</span></code>로 시작한다.</p></li>
<li><p>성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.</p></li>
<li><p>예를 들어 27살 남성은 <code class="docutils literal notranslate"><span class="pre">male27</span></code> 값이 된다.</p></li>
</ol>
</div>

```python
titanic["category2"] = titanic["sex"]+titanic["age"].astype(int).astype(str)
titanic
```

<p class="admonition-title">연습 문제 4.4.7</p>
<p>타이타닉호 승객을 ‘미성년자’, ‘청년’, ‘중년’, ‘장년’, ‘노년’ 나이 그룹으로 나눈다.</p>
<div class="highlight-default notranslate"><div class="highlight"><pre><span></span><span class="n">bins</span> <span class="o">=</span> <span class="p">[</span><span class="mi">1</span><span class="p">,</span> <span class="mi">20</span><span class="p">,</span> <span class="mi">30</span><span class="p">,</span> <span class="mi">50</span><span class="p">,</span> <span class="mi">70</span><span class="p">,</span> <span class="mi">100</span><span class="p">]</span>
<span class="n">labels</span> <span class="o">=</span> <span class="p">[</span><span class="s2">&quot;미성년자&quot;</span><span class="p">,</span> <span class="s2">&quot;청년&quot;</span><span class="p">,</span> <span class="s2">&quot;중년&quot;</span><span class="p">,</span> <span class="s2">&quot;장년&quot;</span><span class="p">,</span> <span class="s2">&quot;노년&quot;</span><span class="p">]</span>
</pre></div>
</div>
<p>그리고 각 나이 그룹의 승객 비율을 구한다. 비율의 전체 합은 1이 되어야 한다.</p>
</div>

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic["percentage"] = pd.cut(titanic.age, bins, labels = labels).astype(str)
titanic[["age","percentage"]]
```

<p class="admonition-title">연습 문제 4.4.8</p>
<p>타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 <code class="docutils literal notranslate"><span class="pre">category3</span></code> 열을 만들어라. <code class="docutils literal notranslate"><span class="pre">category3</span></code> 카테고리는 다음과 같이 정의된다.</p>
<ol class="simple">
<li><p>20살 미만이면 성별에 관계없이 “미성년자”라고 한다.</p></li>
<li><p>20살 이상이면 나이에 따라 “청년”, “중년”, “장년”, “노년”을 구분하고 그 뒤에 성별을 나타내는 “남성”, “여성”을 붙인다.</p></li>
</ol>
</div>

```python
bins = [20,30,50,70,100]
labels = ["청년", "중년", "장년", "노년"]
titanic["category3"] = titanic.apply(lambda x: "미성년자" if x.age < 20
                                    else x.percentage + ":" + "남성" if x.sex == "male"
                                    else x.percentage + ":" + "여성", axis=1)
titanic[["category3"]]
```

<ol class="simple">
<li><p>학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 <code class="docutils literal notranslate"><span class="pre">df_score1</span></code> 을 생성한 후, <code class="docutils literal notranslate"><span class="pre">df_score1.index</span></code> 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. <code class="docutils literal notranslate"><span class="pre">reset_index</span></code> 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 데이터프레임 <code class="docutils literal notranslate"><span class="pre">df_score2</span></code>을 만든다.</p></li>
<li><p>학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 <code class="docutils literal notranslate"><span class="pre">df_score2</span></code>에 <code class="docutils literal notranslate"><span class="pre">set_index</span></code> 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.</p></li>

</div>

```python
df_score1.index = ["학생1","학생2","학생3","학생4","학생5"]
df_score1.index.name = "학생"
df_score1
```

```python
df_score2 = df_score1.reset_index()
df_score2
```

```python
#2
df_score2.set_index("학생")
```

<p class="admonition-title">연습 문제 4.5.2</p>
<p>A 반 학생 5명과 B반 학생 5명의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다.</p>
<ol class="simple">
<li><p>“반”, “번호”, “국어”, “영어”, “수학” 을 열로 가지는 데이터프레임 <code class="docutils literal notranslate"><span class="pre">df_score3</span></code>을 만든다.</p></li>
<li><p><code class="docutils literal notranslate"><span class="pre">df_score3</span></code>을 변형하여 1차 행 인덱스로 “반”을 2차 행 인덱스로 “번호”을 가지는 데이터프레임 <code class="docutils literal notranslate"><span class="pre">df_score4</span></code>을 만든다.</p></li>
<li><p>데이터 프레임 <code class="docutils literal notranslate"><span class="pre">df_score4</span></code>에 각 학생의 평균을 나타내는 행을 오른쪽에 추가한다.</p></li>
<li><p><code class="docutils literal notranslate"><span class="pre">df_score3</span></code>을 변형하여 행 인덱스로 “번호”를, 1차 열 인덱스로 “국어”, “영어”, “수학”을, 2차 열 인덱스로 “반”을 가지는 데이터프레임 <code class="docutils literal notranslate"><span class="pre">df_score5</span></code>을 만든다.</p></li>
<li><p>데이터 프레임 <code class="docutils literal notranslate"><span class="pre">df_score5</span></code>에 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.</p></li>
</ol>
</div>

```python
#1
np.random.seed(0)
df_score3=pd.DataFrame(np.vstack([list('AAAAABBBBB'),
np.array([i for i in range(1,6)]*2)
,np.random.randint(1,101,size=(3,10))]).T
,columns=['반','번호','국어','영어','수학'])
df_score3
```

```python
#2
df_score4=df_score3.set_index(["반","번호"])
df_score4
```

```python
#3
df_score4.loc[:, "평균점수"] = df_score4.mean(axis=1)
df_score4
```

```python
#4
temp = df_score3.set_index(["번호","반"])
df_score5 = temp.unstack("반").sort_index(level=0)
df_score5
```

```python
#5
df_score5.loc["평균",:] = df_score5.mean()
df_score5
```
